```python
import pandas as pd
import numpy as np
```

```python
museums = "Resources/museums copy.csv"
museums_df = pd.read_csv(museums)
museums_df.head()
```

```python
new_columns = ["Museum Name", "Museum Type", "Latitude", "Longitude", "Income", "Revenue"]
museums_df = museums_df.loc[:, new_columns]
museums_df
```

```python
museums_df.dtypes
```

```python
NaN_columns = ["Income", "Revenue"]
museums_df[NaN_columns] = museums_df[NaN_columns].fillna(0.0)
museums_df = museums_df.dropna(how='any')
museums_df.fillna(0.0, inplace=True)
museums_df
```

```python
diff_museum_types = museums_df["Museum Type"].unique()
for museums_types in diff_museum_types:
    print(museums_types)
```

```python
museums_df.to_csv("Resources/museums_df.csv", index=False)
```

```python
trip_advisor = "Resources/tripadvisor_museum_USonly copy.csv"
trip_advisor_df = pd.read_csv(trip_advisor)
trip_advisor_df.head()
```

```python
drop_tripadvisor_columns = ["Unnamed: 0", "Address", "Description", "FeatureCount", "Fee", 
                            "PhoneNum", "TotalThingsToDo", "LengthOfVisit"]
trip_advisor_df.drop(columns=drop_tripadvisor_columns, inplace=True)
trip_advisor_df = trip_advisor_df.rename(columns={"Langtitude": "Longitude"})
trip_advisor_df.dropna(inplace=True)
trip_advisor_df
```

```python
trip_advisor_df.to_csv("Resources/trip_advisor_df.csv", index=False)
```

```python
major_cities = "Resources/uscities.csv"
major_cities_df = pd.read_csv(major_cities)
major_cities_df.head()
```

```python
grab_columns = ["city", "state_name", "lat", "lng", "population"]
major_cities_df = major_cities_df.loc[:, grab_columns]
major_cities_df.head()
```

```python
major_cities_df = major_cities_df.rename(columns={"city" : "City", "state_name" : "State",
                                                  "lat" : "Latitude", "lng" : "Longitude",
                                                  "population" : "Population"})
major_cities_df.head()
```

```python
major_cities_df = major_cities_df.sort_values(by = "Population", ascending=False)
major_cities_df.head(10)
```

```python
major_cities_df = major_cities_df.iloc[:100, :]
major_cities_df.head()
```
